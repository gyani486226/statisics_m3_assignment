```python
# Q9 Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

# The cumulative distribution function of exponential distribution is the differentiation of probability density function up to certain point .

# Formula = ( 1-e^(-lambda * X))
# Where Lambda = average rate >> X = certtain value where we want to find the  cumulative distribution


def cumulative_distribution(lambd, X):
    cumulative = 1 - np.exp(-lambd * X)
    return cumulative


result = cumulative_distribution(.2,6)

print(f"The cumulative distribution value for lambda = 0.2 and X = 6 is:{result:.4f}")
```

```python
#Q11  A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
#     who make a purchase). They collect data from the old and new layouts to compare.

#Apply z-test to find which layout is successful.
# Null hypothesis :- The new website layout does not lead to a higher 
# conversion rate.

#Alternate hypothesis :- The new website layout leads to a higher conversion rate.


import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)
n1 = len(old_layout)

# 70 purchases out of 1000 visitors  

new_layout = np.array([1] * 70 + [0] * 930)
n2 = len(new_layout)

# Applying Z test 
# Step 1 : Calculate  Z acore 

mean_old = np.mean(old_layout)
std_old = np.std(old_layout)
mean_new = np.mean(new_layout)
std_new = np.std(new_layout)

zcritical = 1.96

Zscore = (mean_new-mean_old)/math.sqrt((std_old**2/n1)+(std_new**2/n2))
print(f"Zscore:{Zscore}")
if abs(Zscore) > zcritical:
    print("Reject the null hypothesis:The new website leads to a higher conversion rate")
else:
    print("Fail to reject the null hypothesis:The new website doesn't affect to the conversion")
      
```

```python
#Q12 A tutoring service claims that its program improves students' exam scores. A sample of students who participated in the program was taken, 
# and their scores before and after the program were recorded.
# Take alpha = 0.05

# Data
import numpy as np
import math

# Data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])


before_program_list = before_program.tolist() 
after_program_list = after_program.tolist()
# Null Hypothesis(Ho): The tutoring program does not improve student's exam scores.
#Alternate Hypothesis(H1): The tutoring program improves student's exam scores.

# Sample sizes
n1 = len(before_program)
n2 = len(after_program)

# Means and standard deviations
before_mean = np.mean(before_program)
before_std = statistics.stdev(before_program_list) 
after_mean = np.mean(after_program)
after_std = statistics.stdev(after_program_list) 

# Z-score calculation
zscore = (after_mean - before_mean) / math.sqrt((before_std**2 / n1) + (after_std**2 / n2))
print("Zscore:", zscore)

# Critical value for a 95% confidence level
zcritical = 1.96

# Interpretation
if abs(zscore) > zcritical:
    print("Reject the null hypothesis: The tutoring program improves students' exam scores.")
else:
    print("Fail to reject the null hypothesis: The tutoring program does not improve students' exam scores.")
```

```python
#Q14  A customer service department claims that their average response time is less than 5 minutes. A sample of recent customer interactions was taken, 
# and the response times were recorded.
#Implement z-test to find the claims made by customer service department are true or false

response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Null hypothesis:- Average time is not less than 5 minutes
# Alternate hypothesis:- Average time is less than 5 minutes
# One tail - test

response_mean = np.mean(response_times)
response_std = np.std(response_times, ddof = 1)
n = len(response_times)
average_time = 5
Zscore = (response_mean-average_time)/(response_std/math.sqrt(n))
print("Zscore:", Zscore)

#for one tail test 
z_critical = 1.64

if abs(Zscore) > z_critical:
    print("Reject the null hypothesis: Average time is less than 5 minutes")
else:
    print("Fail to reject the null hypothesis: Average time is not less than 5 minutes")
```

```python
#Q21 A political analyst wants to determine if there is a significant association between age groups and voter preferences
#(Candidate A or Candidate B). 
# They collect data from a sample of 500 voters and classify them into a different age groups and candidate preferences. Perform a Chi-Square test 
# to determine if there is a significant association between age groups and voter preferences.

# Data value
np.random.seed(0)
age_groups = np.random.choice([ "18-30" , "31-50" , '51+', '51+'], size=500)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Null hypothesis = There is no significant relationship between age and voter
# Alternate hypotheses = There is significant relationship between age and voter

# Create a data frame
data = pd.DataFrame({"Age Group": age_groups, "Voter Preference": voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data["Age Group"], data["Voter Preference"])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

alpha = 0.05

if p < alpha:
    print("Reject the null hypothesis : There is a significant relationship between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: There is significant relationship between age groups and voter prefeences")
```
